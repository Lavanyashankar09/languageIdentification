```python
import os
import h5py
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Function to load embeddings from HDF5 file
def load_embeddings(hdf5_file_paths, layer):
    embeddings = []
    labels = []
    
    for hdf5_file_path in hdf5_file_paths:
        with h5py.File(hdf5_file_path, 'r') as hf:
            # Extract embeddings from the specified layer and labels
            embeddings.append(hf[layer][:])
            labels += [label.decode('utf-8') for label in hf['LID'][:]]  # Decode labels
    
    # Stack the embeddings from all files along the second axis (sample dimension)
    embeddings = np.concatenate(embeddings, axis=1)
    return embeddings, labels

# Function to apply PCA and plot the results
def plot_pca(embeddings, labels, layer, output_dir, num_components=2):
    # Standardize the embeddings before applying PCA
    embeddings = StandardScaler().fit_transform(embeddings)
    
    # Apply PCA
    pca = PCA(n_components=num_components)
    principal_components = pca.fit_transform(embeddings)
    
    # Create a scatter plot of the first two principal components
    plt.figure(figsize=(10, 7))
    
    # Map labels to colors (e.g., 'English' -> 0, 'Mandarin' -> 1)
    label_map = {'English': 0, 'Mandarin': 1}
    colors = ['r' if label_map[label] == 0 else 'b' for label in labels]
    
    plt.scatter(principal_components[:, 0], principal_components[:, 1], c=colors, alpha=0.7)
    
    # Add labels and title
    plt.xlabel('Principal Component 1')
    plt.ylabel('Principal Component 2')
    plt.title(f'PCA of {layer} Embeddings')
    
    # Save the plot
    plt.savefig(os.path.join(output_dir, f'pca_plot_{layer}.pdf'))
    plt.close()
    
    print(f'PCA plot for {layer} saved at {os.path.join(output_dir, f"pca_plot_{layer}.pdf")}')

# Main function
def main(hdf5_dir, save_dir_plot, layers):
    # Gather all HDF5 files in the directory
    hdf5_file_paths = [os.path.join(hdf5_dir, f) for f in os.listdir(hdf5_dir) if f.endswith('.h5')]
    
    for layer in layers:
        print(f'Processing {layer}...')
        # Load embeddings for the given layer
        embeddings, labels = load_embeddings(hdf5_file_paths, layer)
        
        # Ensure output directory for the plots exists
        os.makedirs(save_dir_plot, exist_ok=True)
        
        # Plot PCA
        plot_pca(embeddings, labels, layer, save_dir_plot)

if __name__ == "__main__":
    hdf5_dir = '/export/c09/lavanya/languageIdentification/zinglish/large/embeddingLarge/'  # Path to your HDF5 directory
    save_dir_plot = '/export/c09/lavanya/languageIdentification/zinglish/large/plotLarge/'  # Directory to save PCA plots
    
    # List of layers to analyze (e.g., Layer_1, Layer_2, etc.)
    layers = ['Layer_1', 'Layer_2', 'Layer_3', 'Layer_4', 'Layer_5', 'Layer_6']
    
    main(hdf5_dir, save_dir_plot, layers)
```
