```python
import pandas as pd
from jiwer import wer
import ast
import re
import os
# Load CSV files
zipformer_df = pd.read_csv("/export/c09/lavanya/languageIdentification/merlion/whisper_trans.csv")  # Adjust filename as needed
whisper_df = pd.read_csv("/export/c09/lavanya/languageIdentification/merlion/zipformer_trans.csv")
```

```python
whisper_df
```

```python

# Function to extract the file name
def extract_file_name(file_path):
    # Use os.path.basename to get the file name
    return os.path.basename(file_path).strip(',')

# Apply the function to create a new column with just the file names
whisper_df["File Name"] = whisper_df["Audio File"].apply(extract_file_name)
```

```python

# Function to clean the transcription
def clean_transcription(transcription):
    if isinstance(transcription, str):
        # Remove square brackets and single quotes
        transcription = re.sub(r"[\[\]']", "", transcription)
        # Remove punctuation
        transcription = re.sub(r"[^\w\s]", "", transcription)
        return transcription
    else:
        # Return the transcription as is if it's not a string
        return transcription
```

```python
# Apply the cleaning function to the "Transcription (List of Words)" column
zipformer_df["Transcription Zipformer"] = zipformer_df["Transcription (List of Words)"].apply(clean_transcription)
```

```python
whisper_df["Transcription whisper"] = whisper_df["Transcription"].apply(clean_transcription)
```

```python
zipformer_df
```

```python
# Make a copy of the DataFrame
z = zipformer_df.copy()

# Drop specific columns from the copy
columns_to_drop = ['Transcription (Uppercase)', 'Transcription (List of Words)']
z1 = z.drop(columns=columns_to_drop)
```

```python
z1.head()
```

```python
# Rename a single column
z1 = z1.rename(columns={'Audio File': 'file_name'})
z1
```

```python

w = whisper_df.copy()

# Drop specific columns from the copy
columns_to_drop = ['Transcription', 'Audio File']
w1 = w.drop(columns=columns_to_drop)
# Rename a single column
w1 = w1.rename(columns={'File Name': 'file_name'})
```

```python
# Merge the DataFrames on the 'File Name' column
merged_df = pd.merge(z1, w1, on='file_name', how='inner')

# Display the merged DataFrame
print(merged_df)

# Optionally, save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_output.csv', index=False)
```

```python
merged_df
```

```python
import numpy as np

wer_results = []
for i in range(len(merged_df)):
    reference = merged_df.loc[i, "Transcription Zipformer"]
    hypothesis = merged_df.loc[i, "Transcription whisper"]

    # Check if either reference or hypothesis is NaN or empty
    if pd.isna(reference) or pd.isna(hypothesis) or not reference or not hypothesis:
       # print(f"Skipping Row {i} due to NaN or empty reference or hypothesis.")
        wer_results.append(np.nan)  # Append NaN for skipped rows
        continue

    error_rate = wer(reference, hypothesis)
    wer_results.append(error_rate)

# Add WER results as a new column in the DataFrame
merged_df["WER"] = wer_results

# Optionally, print the DataFrame to check the new column
print(merged_df)
```

```python
# Save the DataFrame to a CSV file
merged_df.to_csv('merged_with_wer_results.csv', index=False)

# Optionally, print the DataFrame to check the new column
print(merged_df)
```

```python
# Calculate the average of the WER column
avg_wer = merged_df["WER"].mean()

# Print the average WER
print(f"Average WER: {avg_wer:.2f}")
```
