```python
import os
common_path = "/export/c09/lavanya/seame_data/data"
conv_txt_path = common_path + '/conversation/transcript/phaseII'
txt_files = [f for f in os.listdir(conv_txt_path) if f.endswith('.txt')]
print(f"Number of .txt files: {len(txt_files)}")

# Initialize counters for each language
language_counts = {
    'EN': 0,
    'ZH': 0,
    'CS': 0
}

# Initialize total duration counters for each language in milliseconds
language_durations = {
    'EN': 0,
    'ZH': 0,
    'CS': 0
}

# Function to convert milliseconds to hours, minutes, and seconds
def ms_to_time(ms):
    total_seconds = ms / 1000
    hours = total_seconds // 3600
    minutes = (total_seconds % 3600) // 60
    seconds = total_seconds % 60
    return int(hours), int(minutes), int(seconds)

# Iterate through each .txt file and count language occurrences and calculate durations
for file in txt_files:
    with open(os.path.join(conv_txt_path, file), 'r', encoding='utf-8') as transcript_file:
        for line in transcript_file:
            # Split the line by tab and get the language code (assuming the 4th column is the language code)
            parts = line.strip().split('\t')
            if len(parts) > 3:  # Ensure the line has enough columns
                start_time = int(parts[1])  # Start time in milliseconds
                end_time = int(parts[2])  # End time in milliseconds
                language_code = parts[3]  # Get the language code

                # Calculate duration for the current segment
                duration = end_time - start_time

                # Add to language duration if the language code is recognized
                if language_code in language_durations:
                    language_durations[language_code] += duration
                    language_counts[language_code] += 1  # Increment the count of segments for that language

# Print the counts and durations for each language
print("Language counts and durations:")
for language, count in language_counts.items():
    duration_ms = language_durations[language]
    hours, minutes, seconds = ms_to_time(duration_ms)
    print(f"{language}: {count} segments, Total duration: {hours}h {minutes}m {seconds}s")
```

```python
from jiwer import wer

def calculate_wer_from_rows(data):
    """
    Calculate WER for each row in the dataset.

    Parameters:
    data (list of tuples): Each tuple contains the reference (ground truth) and hypothesis.

    Returns:
    list of tuples: Each tuple contains the row index, reference, hypothesis, and WER percentage.
    """
    results = []

    for i, (reference, hypothesis) in enumerate(data):
        # Clean and preprocess hypothesis to convert string representation of list to sentence
        hypothesis_cleaned = ' '.join(eval(hypothesis))  # Evaluate string list safely
        error_rate = wer(reference, hypothesis_cleaned) * 100
        results.append((i + 1, reference, hypothesis_cleaned, error_rate))

    return results


# Sample data (extracted from your provided dataset)
data = [
    ("[ah] final saturday", "['OUR', 'FINAL', 'SATURDAY']"),
    ("八十六  加", "['八', '十', '六', '家']"),
    ("and hello","and hello"),
    ("就  八十  我  拿  了 一  三  九  [啊]", "['就', '巴', '士', '我', '拿', '一', '三', '九', '娃']"),
    ("就  八十","['就', '八', '十']")
]

# Calculate WER for each row
wer_results = calculate_wer_from_rows(data)

# Display results
for row in wer_results:
    print(f"Row {row[0]}:")
    print(f"Reference: {row[1]}")
    print(f"Hypothesis: {row[2]}")
    print(f"WER: {row[3]:.2f}%\n")
```
