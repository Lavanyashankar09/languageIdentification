```python
import pandas as pd
import librosa
import soundfile as SF
import pickle
with open('/export/c09/lavanya/languageIdentification/test_pickles/segments.pkl', 'rb') as f:
    seg_df = pickle.load(f)
```

```python
# Define the old and new path prefixes
old_prefix = '/export/c09/lavanya/languageIdentification/segments/'
new_prefix = '/export/c09/lavanya/languageIdentification/test_segments_40/'

# Update the 'segmented_audio' column
seg_df['segmented_audio'] = seg_df['segmented_audio'].replace(to_replace=old_prefix, value=new_prefix, regex=True)
seg_df['segmented_audio'][0]
```

```python
# Get unique values and count them
unique_values = seg_df['segmented_audio'].unique()
unique_count = len(unique_values)

# Print the count of unique values
print(f"Number of unique values in 'segmented_audio': {unique_count}")
```

```python
test_embeddings_base = '/export/c09/lavanya/languageIdentification/test_embedding_40/'

# Define a function to generate the embedding path
def get_embedding_path(segment_path):
    # Extract the file name from the segment path
    file_name = segment_path.split('/')[-1]
    # Replace the base path with the embedding path
    embedding_path = test_embeddings_base + f'emb_{file_name.replace(".wav", "")}.pt'
    return embedding_path

# Apply the function to create the 'embedding' column
seg_df['embedding'] = seg_df['segmented_audio'].apply(get_embedding_path)
```
