```python
import h5py
import numpy as np

# Path to the HDF5 file
path = "/export/c09/lavanya/languageIdentification/zinglish/embeddingSmall/embed_TTS_P42566TT_VCST_ECxxx_02_AO_25789991_v001_R007_CRR_MERLIon-CCS_segment_56_Mandarin.h5"

# Open the HDF5 file in read mode
with h5py.File(path, 'r') as hf:
    print("type(hf):", type(hf))
    print("hf.len():", len(hf))
    print("hf.keys():", hf.keys())
    print("hf.values():", hf.values())
    print("hf.items():", hf.items())
    print("hf.attrs:", hf.attrs)
    
```

```python
# Open the HDF5 file in read mode
with h5py.File(path, 'r') as hf:
   
    for k in hf.keys():
        print("k is ", k)
    for v in hf.values():
        print("v is ",v)
```

Explanation:
v[...]: This retrieves the full dataset as a NumPy array for non-scalar datasets.
v[()]: This retrieves scalar values, such as the language tags (datasets with no shape).

```python
with h5py.File(path, 'r') as hf:
    for k, v in hf.items():
        print(f"Dataset {k}:")
        # Check if the dataset is scalar or an array
        if v.shape == ():  # Scalar value, such as language_tag
            print(v[()])  # Prints scalar value
        else:
            print(v[...])  # Prints the full dataset (as a NumPy array)
        print('-' * 50)  # Separator for clarity
```

```python
with h5py.File(path, 'r') as hf:
    for k in hf.keys():
        print(f"Dataset {k}:")
        v = hf[k]  # Get the dataset or attribute associated with key `k`
        
        # Check if the dataset is scalar or an array
        if v.shape == ():  # Scalar dataset (e.g., language tags)
            print(v[()])  # Print scalar value
        else:  # Non-scalar dataset (e.g., embeddings)
            print(v[...])  # Print the full dataset as a NumPy array
            
        print('-' * 50)  # Separator for clarity
```
